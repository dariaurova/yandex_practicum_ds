### Эффективность обогощения

Проверим, что эффективность обогащения рассчитана правильно. 
Вычислим её на обучающей выборке для признака rougher.output.recovery. Найдем MAE между расчётами и значением признака. 


C — доля золота в концентрате после флотации/очистки;
F — доля золота в сырье/концентрате до флотации/очистки;
T — доля золота в отвальных хвостах после флотации/очистки.
Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. Причём важен не только финальный продукт, но и черновой концентрат.


```python
f = data_train['rougher.input.feed_au']
t = data_train['rougher.output.tail_au']
c = data_train['rougher.output.concentrate_au']
recovery = (c*(f - t))/(f*(c - t))*100

mae_recovery = mean_absolute_error(data_train['rougher.output.recovery'], recovery)
print('Средняя абсолютная ошибка: ',mae_recovery)
print('Среднее эффективности обогощения по расчету: ',recovery.mean())
print('Среднее эффективности обогощения по выборке: ',data_train['rougher.output.recovery'].mean())
```

Получаем, что средняя абсолютная ошибка близка к нулю, что значит, что результаты расчета и значения признака практически идентичны. Значит эффективность обогощения рассчитана правильно.

```python
# Найдем распределение средних
def boots(sries):
    state = np.random.RandomState(12345)
    values = []
    for i in range(1000):
        sries_subsample = sries.sample(n = 500, replace=True, random_state=state)
        values.append(sries_subsample.mean())
    values = pd.Series(values)
    return values
```

```python
fig, ax = plt.subplots(figsize=(10,5), constrained_layout=True)
plt.hist(boots(conc_ag['rougher.output.concentrate_ag']), label='продукт флотации')
plt.hist(boots(conc_ag['primary_cleaner.output.concentrate_ag']), label='продукт очистки')
plt.hist(boots(conc_ag['final.output.concentrate_ag']), label='финальный продукт')
ax.legend(loc='best')
ax.set_xlabel('Концентрация')
ax.set_ylabel('Частота')
ax.set_title('Распределение средней концентрации Ag для трех этапов');
```

### Распределения размеров гранул сырья
Сравним распределения размеров гранул сырья на обучающей и тестовой выборках. Если распределения сильно отличаются друг от друга, оценка модели будет неправильной.
Исследуем суммарную концентрацию всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах.

Для начала посмотрим на средние значения размеров гранул сырья на тестовой и обучающей выборке, а также на размер данных на этих выборках

```python


print('Средний размер гранул для обучающей выборки',data_train['rougher.input.feed_size'].mean())
print('Средний размер гранул для тестовой выборки',data_test['rougher.input.feed_size'].mean())
print('Количество значений в обучающей выборке',data_train['rougher.input.feed_size'].count())
print('Количество значений в тестовой выборке',data_test['rougher.input.feed_size'].count())
```

Построим графики распрделения размеров гранул для двух выборок

```python
plt.xlabel('Размер зерна')
plt.ylabel('Плотность')
plt.title('Гистограмма и плотность распределения размера зерна');
ax = sns.distplot(data_train['rougher.input.feed_size'], hist=True,kde=True, bins=int(1000), label='train size');
ax = sns.distplot(data_test['rougher.input.feed_size'], hist=True,kde=True, bins=int(1000), label='test size');
ax = plt.axvline(x=data_train['rougher.input.feed_size'].mean(), color='darkblue', label='train mean size')

ax = plt.axvline(x=data_test['rougher.input.feed_size'].mean(), color = 'darkgreen', label='train test size')
plt.xlim([10, 120])
plt.legend();
```

На графиках распределения мы можем заметить, что гистограммы и плотность распредеделния на тестовой и тренировочной выборке имеют очень похожую форму. Также можем заметить, что тестовая выборка смещена влево относительно тренировочной. Получаем, что средний размер зерна на тестовой выборке равен 55.9, что меньше среднего размера для обучающей выборки (60.2).

Получаем, что среднее на тестовой выборке и на обучающей различаются на 4 единицы. Проверим насколько это большое различие с помощью t теста. Проверим гипотезу о том, что распределения не сильно отличаются друг от друга.

H_0: 'Распределения размеров гранул сырья на обучающей и тестовой выборках не сильно различаются друг от друга'

```python

alpha = .05 # критический уровень статистической значимости

results = st.ttest_ind(
    data_train['rougher.input.feed_size'],
    data_test['rougher.input.feed_size'], equal_var=False
)

print('p-значение:', results.pvalue)

if results.pvalue < alpha: 
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

В ходе проверки гипотезы при статичстическом уровне значимости alha = 5%, нулевая гипотеза о том, что распределения размеров гранул сырья на обучающей и тестовой выборках не сильно различаются друг от друга, была отвергнута. При этом на графике распределения видно, что распределения очень близки друг другу, возможно ttest для наших данных является слишком точным инструментом.

### Функция для вычисления итогового sMAPE

```python
def smape(true_values, pred):
    return 1/len(pred) * np.sum(2 * np.abs(true_values-pred) / (np.abs(pred) + np.abs(true_values))*100)
```

Итоговое sMAPE


```python
def final_smape(best_smape_r, best_smape_f):
    fin_smape = (0.25*best_smape_r)+(0.75*best_smape_f)
    return fin_smape
```

#### Прогнозирование  rougher.output.recovery

```python
from catboost import CatBoostRegressor

class MyMetric:
    @staticmethod
    def get_profit(y_true, y_pred):
        #y_pred = expit(y_pred).astype(int)
        #y_true = y_true.astype(int)
        return (1/len(y_pred) * np.sum(2 * np.abs(y_true-y_pred) / (np.abs(y_pred) + np.abs(y_true))*100))
    
    def is_max_optimal(self):
        return True # greater is better

    def evaluate(self, approxes, target, weight):            
        assert len(approxes) == 1
        assert len(target) == len(approxes[0])
        y_true = np.array(target).astype(int)
        approx = approxes[0]
        score = self.get_profit(y_true, approx)
        return score, 1

    def get_final_error(self, error, weight):
        return error
cat = CatBoostRegressor(eval_metric=MyMetric())
cat.fit(features_train_rougher, target_train_rougher)
```
